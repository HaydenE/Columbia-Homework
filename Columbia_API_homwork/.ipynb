```python
latitudes = []
longitudes = []
for x in range(500):
    latitudes.append(random.randint(-23,23))
    longitudes.append(random.randint(-180,180))
```

```python
city_list = []
for counter, value in enumerate(latitudes):
    coordinates = (value, longitudes[counter])
    city = citipy.nearest_city(coordinates[0],coordinates[1])
    city_name = city.city_name
    c_code = city.country_code
    city_list.append([city_name,c_code,latitudes[counter]])

#city = citipy.nearest_city(-34,44)
#city.city_name
```

```python
# Specify the URL
data = []
for city in city_list:
    url = "http://api.openweathermap.org/data/2.5/weather?q="+city[0]+"&APPID="+api
    response = requests.get(url)
    data.append(response.json())
```

```python
weather_info = []
for datum in data:
    try:
        weather_info.append([datum['name'],datum['main']['temp'],datum['main']['humidity'],datum['clouds']['all'],datum['wind']['speed']])
    except:
        print("No city found!")
```

```python
city_names = []
city_lat = []
city_temp = []
city_hum = []
city_clouds = []
city_wind = []

for datum in data:
    try:
        city_lat.append(datum['coord']['lat'])
        city_names.append(datum['name'])
        city_temp.append(datum['main']['temp'])
        city_hum.append(datum['main']['humidity'])
        city_clouds.append(datum['clouds']['all'])
        city_wind.append(datum['wind']['speed'])
    except:
        print('nope')
```

```python
result = set(zip(city_names, city_lat, city_temp, city_hum, city_clouds, city_wind))
df = pd.DataFrame(result)
```
