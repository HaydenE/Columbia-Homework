```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(city_data_df, ride_data_df, how="right", on=["city", "city"])

# Display the data table for preview
pyber_df.head(5000)
```

```python
city_grouped = pyber_df.groupby('city', sort=False, observed=True)
city_grouped_mean = city_grouped.mean()
city_grouped_count = city_grouped.count()

city_mean_regions = pd.merge(city_data_df, city_grouped_mean, how="right", on=["city", "city"])
city_mean_and_count = pd.merge(city_mean_regions, city_grouped_count, how="right", on= ["city", "city"])
#city_mean_regions
city_mean_and_count
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
colors = np.where(city_mean_and_count["type_x"]== "Urban",'red', np.where(city_mean_and_count["type_x"] == "Suburban", "blue", "yellow"))
plt.style.use('seaborn-bright')
# Build the scatter plots for each city types
plt.scatter(x= city_mean_and_count['driver_count'],y=city_mean_and_count['fare_x'],c=colors,s= 4 * city_mean_and_count["driver_count_y"],edgecolor='black',alpha=0.5)
# Incorporate the other graph properties

# Create a legend
plt.title("Pybar Ride Share Chart")
plt.xlabel("Total Number of Rides per City")
plt.ylabel("Average Fare ($)")
# Incorporate a text label regarding circle size
plt.text(45, 30, "Note: dot size correlates with number of drivers per city", horizontalalignment='left', size='medium', color='black', weight='semibold')

# Save Figure
plt.savefig("../PybarRideScatter.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_fares_grouped = pyber_df.groupby('type')
sum_city_fares = city_fares_grouped.sum()
# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = sum_city_fares['fare']

# The colors of each section of the pie chart
colors = ["red", "orange", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0, 0)

plt.pie(sizes, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
count_city_fares = city_fares_grouped.count()
# Build Pie Chart
sizes = count_city_fares['fare']

plt.pie(sizes, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Save Figure
sum_city_fares
```

```python
# Show Figure
sum_city_fares
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
sizes = sum_city_fares['driver_count'].unique()

# Build Pie Charts
driver_percent_pie = plt.pie(sizes, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
# Save Figure
driver_percent_pie
```
