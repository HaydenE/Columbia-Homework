```python
# Dependencies and Setup
import pandas as pd
pd.set_option('display.max_columns', 50)

# File to Load (Remember to Change These)
school_data_to_load = R"C:\Users\hayde\Col_Class\colnyc201907data3\Homework\04-Pandas\Instructions\PyCitySchools\Resources\schools_complete.csv"
student_data_to_load = R"C:\Users\hayde\Col_Class\colnyc201907data3\Homework\04-Pandas\Instructions\PyCitySchools\Resources\students_complete.csv"

```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
sdc = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
pd.DataFrame(sdc)
```

```python
# High Level SnapShot Data
stats = sdc.describe([.70])
total_students = stats.iloc[ 0, 0]
total_budget = stats.iloc[0,5]

average_math_score = stats.iloc[1,2]
average_reading_score = stats.iloc[1,1]

percent_passing_math = stats.iloc[5,2]
percent_passing_reading = stats.iloc[5,1]
overall_passing_rate = (percent_passing_reading + percent_passing_math) / 2

total_school_number =  len(school_data)


#print(student_count)
print(stats)
#print(stats['reading_score'])
```

```python
High_level_snapshot = {
    'Total Schools' : [total_school_number],
    'Total Students' : [total_students],
    'Total Budget' : [total_budget],
    'Average Math Score' : [average_math_score],
    'Average Reading Score' : [average_reading_score],
    '% Passing Math' : [percent_passing_math],
    '% Passing Reading' : [percent_passing_math],
    'Overall Passing Rate' : [overall_passing_rate]
}
High_level_snapshot_df = pd.DataFrame(High_level_snapshot)
High_level_snapshot_df
```

```python
#School Summary
grouped_school = sdc.groupby('school_name', sort=False)

#means
school_means = grouped_school.mean()
school_means_final = pd.merge(school_means['reading_score'], school_means['math_score'],right_index = True, left_index=True)

#adding means
school_data = pd.merge(school_data, school_means_final, on='school_name', right_index = True, left_index=False)

#per student budget
school_data['budget_per_student']= school_data['budget']/school_data['size']


#% Passing Rates
schools_stats = grouped_school.describe(percentiles=[.70])
schools_reading = schools_stats['reading_score']
schools_math = schools_stats['math_score']
schools_reading_passing_rate = schools_reading['70%']
schools_math_passing_rate = schools_math['70%']

#adding percentage passing rates
school_data_all = pd.merge(school_data, schools_reading_passing_rate, on='school_name' ,right_index = True, left_index=False)
school_data_all = pd.merge(school_data_all, schools_math_passing_rate, on='school_name' ,right_index = True, left_index=False)


#rename columns
school_data_all = school_data_all.rename(columns={'70%_x' : '% Passing Reading', '70%_y' : '% Passing Math'})

#overall Passing Rates
S_overall_passing_rate = (school_data_all['% Passing Reading'] + school_data_all['% Passing Math'])/2
S_overall_passing_rate = pd.DataFrame(S_overall_passing_rate)
school_data_all = pd.merge(school_data_all, S_overall_passing_rate,right_index = True, left_index=True)

#rename again (oop)
school_data_all = school_data_all.rename(columns={0 : 'Overall Passing Rate'})
```

```python
#data by $ per student
bins = [0,597.25,616.5,635.75,700]
labels = ['poor','medium-low','medium-high','high']
school_data_all['binned_by_$'] = pd.cut(school_data_all['budget_per_student'], bins = bins, labels = labels )

school_data_all.iloc[:,[1,2,5,6,8,9,11]]
```

```python
bins = [0,500,1500,3000,5000]
labels = ['very small','small', 'medium', 'large']
school_data_all['Binned_by_school_size'] = pd.cut(school_data_all['size'], bins = bins, labels = labels)
school_data_all.iloc[:,[1,5,6,8,9,12]]
```
